### Could one do it without plotting?

```python
selection = data.groupby((data.date.apply(lambda x: x.weekday()))).mean()
selection.index = [calendar.day_name[i] for i in selection.index]
selection
```

```python
selection = data.groupby((data.date.apply(lambda x: x.month))).mean()
selection.index = [calendar.month_name[x] for x in selection.index]
selection
```

```python
seasons = {1: 'winter', 2: 'spring', 3: 'spring', 4: 'spring', 5:'spring', 6: 'summer',
          7: 'summer', 8: 'summer', 9: 'autumn', 10: 'autumn', 11: 'winter', 12: 'winter'}
```

```python
selection = data.groupby([data.date.apply(lambda x: seasons[x.month]), 
                          data.date.apply(lambda x: calendar.day_name[x.weekday()])], sort = False).mean()
selection.index.names = ['season', 'weekday']
#remove selection here 
selection = selection.loc[(slice(None),['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']), :]
```

```python
selection['coef'] = selection.sunshine_duration/(selection.precipitation_height**2*selection.cloud_cover)
selection.sort_values(['season', 'coef'], ascending=[1,0]) ### shall work for 0.25
```

```python
def some_func(x):       
    return x.coef['autumn']*1.5 + x.coef['spring'] + x.coef['summer']*1.5 + x.coef['winter']*4

result = selection.groupby('weekday').apply(some_func)
result.index.droplevel(0)
result
```
