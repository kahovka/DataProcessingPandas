#### Note the difference between null and none and NaN types! 
There is no null in python. Instead there is a None type or NaN 

None is a null value or no value at all. Not 0, not false, not empy. None is a datatype of its own (NoneType Object) and only None can be None.
NaN  - not a number - null type for numpy.

```python
data_temp = data.replace([-999], [None])
data_temp.head()
```

```python
data_temp.info()
```

### None and NaN are essentially equivalent in pandas in terms of treatment: None is inhereted from python, NaN - from numpy, but both are treated as "not assigned"

```python
data[data.wind_speed == None]['Date'].apply(lambda x: x.year).value_counts().sort_index() #try on data_temp
```

```python
data[data.wind_speed.isna()]['Date'].apply(lambda x: x.year).value_counts().sort_index()
```

```python
data[data.sunshine_duration.isna()]['Date'].apply(lambda x: x.year).value_counts().sort_index()
```
